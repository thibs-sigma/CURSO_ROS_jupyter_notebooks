##  ROS Services

- Los servicios son otra forma en que los nodos pueden comunicarse entre sí. Los servicios permiten a los nodos enviar una solicitud (*Request*) y recibir una respuesta (*Response*) 

### rosservice
- `rosservice` puede conectarse fácilmente a la estructura cliente/servicio de ROS con servicios. 

- `rosservice` tiene muchos comandos que pueden ser utilizados en los servicios. Intenta:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> | `rosservice -h` 

- Eso da la respuesta siguiente:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/utilities-terminal.svg"> | <img src="IMAGES/rosservice_h.png">

#### rosservice list

- Suponiendo que el `turtlesim` sigue ejecutando, intenta el comando siguiente:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosservice list`

- Verás algo parecido a eso:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/utilities-terminal.svg"> | <img src="IMAGES/rosservice_list.png">

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/notice.svg"> | El comando `list` nos muestra que el nodo `turtlesim` brinda 9 servicios, `reset`, `clear`, `spawn`, `kill`, `turtle1/set_pen`, `/turtle1/teleport_absolute`, `/turtle1/teleport_relative`, `turtlesim/get_loggers`, y `turtlesim/set_logger_level`. <br> También hay dos servicios relacionados con el nodo `rosout` separado: `/rosout/get_loggers` y `/rosout/set_logger_level`

#### rosservice type

- Vamos a ver con más detalles el servicio `clear` usando el comando `rosservice type`. Se usa de la manera siguiente:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosservice type [service]`

- Vemos de qué tipo es el servicio `clear`:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosservice type /clear`

- Da la respuesta siguiente:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/utilities-terminal.svg"> | <img src="IMAGES/rosservice_type.png">

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/notice.svg"> | Este servicio está vacío, es decir, cuando se realiza la llamada de servicio no recibe argumentos (es decir, no envía datos al realizar una solicitud y no recibe datos al recibir una respuesta).

#### rosservice call

- Se usa de la manera siguiente:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosservice call [service] [args]`

- Llamamos el service `clear`, sin argumentos porque el servicio es de tipo `empty`:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosservice call /clear`

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/notice.svg"> | Como esperado (con el nombre del *Service*), limpia el fondo del `turtlesim_node`.

- Tomamos como ejemplo un otro *Service* que necesita argumentos, `/spawn`:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosservice type /spawn | rossrv show`

- Da la respuesta siguiente:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/utilities-terminal.svg"> | <img src="IMAGES/rosservice_type_show.png">

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/notice.svg"> | Este servicio nos permite generar una nueva tortuga en un lugar y orientación determinados. El campo de nombre es opcional, así que no le demos un nombre a nuestra nueva tortuga y dejemos que `turtlesim` cree uno para nosotros. 

- Intenta el comando siguiente:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosservice call /spawn 2 2 0.2 ""`

- El comando `service call` responde con el nombre de la tortuga recién creada, y con una nueva tortuga en el `turtlesim`:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/utilities-terminal.svg"> | <img src="IMAGES/rosservice_call.png">
