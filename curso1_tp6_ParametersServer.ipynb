##  ROS Parameters

- `rosparam` te permite almacenar y manipular datos en el *ROS Parameters Server*. 

- El *ROS Parameters Server* puede almacenar enteros, flotadores, booleanos, diccionarios y listas. 

- `rosparam` utiliza el lenguaje de marcado `YAML` para la sintaxis. En casos simples, `YAML` se ve muy natural: 
    - `1` es un entero, 
    - `1.0` es un flotador, 
    - `uno` es una cadena, 
    - `true` es un booleano,
    - `[1, 2, 3]` es una lista de enteros, 
    - `{a: b, c: d}` es un diccionario.

#### rosparam dump and rosparam load

- Esos comandos permiten guardar los parametros en un archivo y recargarlos más tarde.

- Se usa de la manera siguiente:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosparam dump [file_name] [namespace]` <br> `$ rosparam load [file_name] [namespace]`

- Intenta escribir parametros en el archivo `params.yaml`:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosparam dump params.yaml`

- Abrelo en Gedit (por defecto, podes encontrar el archivo en tu carpeta personal `/home`):

<img src="IMAGES/params_yaml.png">

- Y luego, cargalo de nuevo:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosparam load params.yaml`
