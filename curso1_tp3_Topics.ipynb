## ROS Topics

- El nodo `turtlesim_node` y el nodo `turtle_teleop_key` comunican entre sí a través de un *ROS Topic*. 

- `turtle_teleop_key` publica las presionas de las teclas en un *Topic*, mientras que `turtlesim` se suscribe al mismo *Topic* para recibir las presionas de las teclas. 

- Vamos a usar `rqt_graph`, que muestra los nodos y *Topics* que se están ejecutando actualmente. 

### rqt_graph

- `rqt_graph` crea un gráfico dinámico de lo que sucede en el sistema. 

- Se ejecuta gracias a:  

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosrun rqt_graph rqt_graph`

- Verás algo parecido:

<img src="IMAGES/turtlesim_graph.png">

>| []() |  []() |
|---|---| 
| <img src="IMAGES/objects/notice.svg"> | Si coloca el ratón sobre el *Topic* `/turtle1/command_velocity` resaltará los nodos ROS (aquí azul y verde) y los *Topics* (aquí rojo). Como puedes ver, los nodos `turtlesim_node` y `turtle_teleop_key` se están comunicando sobre el *Topic* `/turtle1/command_velocity`.

### rostopic echo

- `rostopic echo` muestra los datos publicados sobre un *Topic*

- Se usa como eso:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rostopic echo [topic]`

- Por ejemplo, vamos a analizar los datos de velocidad publicados por el nodo `turtle_teleop_key`:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rostopic echo /turtle1/cmd_vel`

>| []() |  []() |
|---|---| 
| <img src="IMAGES/objects/notice.svg"> | Probablemente nada va a occurir porque no se están publicando datos sobre el *Topic*.<br> Intenta presionar las teclas de flechas para que  `turtle_teleop_key` publique los datos.

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/important-red.svg"> | **Recuerda que si la tortuga no se está moviendo, tienes que volver a seleccionar el terminal `turtle_teleop_key`**

- Ahora debería ver lo siguiente cuando presione la tecla de arriba: 

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/utilities-terminal.svg"> | <img src="IMAGES/teleop_topic.png">

- Ahora regresamos a `rqt_graph`. Presione el botón de actualización en la parte superior izquierda para mostrar el nuevo nodo. 

<img src="IMAGES/turtlesim_graph2.png">

- `rostopic echo`, mostrado aquí en rojo, ahora está suscrito al *Topic* `turtle1/command_velocity`
