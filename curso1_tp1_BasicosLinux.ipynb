## ¿Qué es Jupyter?
Jupyter es un entorno de trabajo interactivo que permite:
- desarrollar código en Python esentialmente, pero también con otros lenguajes, de manera dinámica
- integrar en un mismo documento tanto bloques de código como texto, gráficas o imágenes

En este curso, este entorno va a ser muy util para illustrar los conceptos teóricos de ROS con prácticas sencillas, si perder tiempo en copiar lineas de código (con los errores que lo accompañan).

### Manipulaciones con el Terminal
- Abre el Terminal. Deberías obtener una pantalla parecida a esa:
<img src="IMAGES/terminal.png">

- Lo primero que verás en la mayoría de los emuladores de terminal es el texto puesto antes de cada línea que escribas en el terminal, que consta de 3 partes:
    1. El nombre de usuario actual de la sesión 
    > en nuestro caso, *"user"*
    2. El nombre de "host", o nombre del dispositivo en la red, actual (después del símbolo @)
    > en nuestro caso *"ubuntu"*
    3. La ubicación actual (después de la :)
    > en nuestro caso aquí es ~, que es un símbolo que representa el directorio raíz del usuario

- Empezamos con un comando muy sencillo. Copia la linea siguiente:
```bash
$ cd /home
```
    
>| []() |  []() |
|---|---|
| <img src="IMAGES/objects/important-red.svg"> | *No se debe copiar el simbolo $, así que solo es una convención en Linux para representar una nueva linéa en el Terminal* |
    
    
> El comando anterior consiste en dos cosas: El comando base cd, y el argumento que le pasas que es /home/. Después de escribir el comando y presionar el botón "Enter" en su teclado, notará que la carpeta actual de su sesión shell fue cambiada a la carpeta /home/, lo que significa que /home/ se está convirtiendo en su directorio de trabajo actual.


- Luego, deseamos listar los elementos (archivos y carpetas) presentes en la carpeta de trabajo actual:
```bash
$ ls
```
    
>| []() |  []() |
|---|---| 
| <img src="IMAGES/objects/notice.svg"> | También puede pasar opciones (u opciones múltiples) a los comandos. Por ejemplo, la opción -a, si se agrega, mostrará los archivos ocultos (cuyos nombres comienzan con un .) | 

- Al final, vamos a:
    1. Crear una carpeta;
    2. Mover hacia esa carpeta;
    3. Listar los archivos presentes adentro (debería ser vacio);
    4. Regresar a la carpeta `home`;
    5. Eliminar la carpeta
    
```bash
$ mkdir Test
$ cd Test
$ ls -a
$ cd ..
$ rm Test
```
    
>| []() |  []() |
|---|---|
| <img src="IMAGES/objects/important-red.svg"> | *Es importante exectuar cada linea a la vez, y no copiar todo en el Terminal...* |

### Herramientas del Sistema de Archivos

- El código está repartido en muchos paquetes de ROS (veremos eso en más detalles más adelante). 


- Navegar con herramientas de línea de comandos como `ls` y `cd` puede ser muy fastidioso. Así, ROS proporciona herramientas para ayudarte. 

#### rospack

- `rospack` te permite obtener información sobre los paquetes. En este tutorial, sólo vamos a cubrir la opción `find`, que devuelve la ruta (`path`) al paquete. 


- Se usa de esa manera:
```bash
$ rospack find [package_name]
```

- Intenta con ese ejemplo:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rospack find roscpp`

- Eso retornaría: `/opt/ros/kinetic/share/roscpp`


#### roscd

- `roscd` te permite cambiar el directorio (`cd`) directamente a un paquete. 

- Se usa de esa manera:
```bash
$ roscd [locationname[/subdir]]
```

- Intenta con ese ejemplo:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ roscd roscpp`

- Verifica que el directorio cambió, gracias al terminal, por: `/opt/ros/kinetic/share/roscpp`

- `roscd` también puede moverse a un subdirectorio de un paquete. Intenta:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ roscd roscpp/cmake`

- Deberías obtener: `/opt/ros/kinetic/share/roscpp/cmake`


#### rosls

- `rosls` te permite listar (`ls`) directamente los archivos y elementos de un Package por su nombre, en lugar de por su ruta (Path).

- Se usa de esa manera:
```bash
$ rosls [locationname[/subdir]]
```

- Intenta con ese ejemplo:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ rosls roscpp_tutorials`

- Eso retornaría los archivos y carpetas del Package `roscpp_tutorials`:
`cmake launch package.xml  srv`


#### Tab cumplimiento

- Puede resultar muy fastidioso (por experiencia, puedo lo garantizarte...) escribir el nombre completo de un paquete. En el ejemplo anterior, `roscpp_tutorials` es un nombre bastante largo.

- Por suerte, algunas herramientas de ROS soportan la cumplimentación con la tecla `TAB`. Intenta:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ roscd roscpp_tut<<< ahora pulse la tecla TAB >>>`

- Después de presionar la tecla `TAB`, la línea de comandos debe rellenar el resto:
`$ roscd roscpp_tutorials/`

>| []() |  []() |
|---|---| 
| <img src="IMAGES/objects/notice.svg"> | Esto funciona porque `roscpp_tutorials` es actualmente el único paquete de ROS que comienza con `roscpp_tut` | 

- Ahora intenta escribir:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |`$ roscd tur<<<<< ahora pulse la tecla TAB >>>>`

- Después de pulsar la tecla TAB, la línea de comandos debe rellenarse tanto como sea posible: `$ roscd turtle`

- Sin embargo, en este caso hay múltiples paquetes que comienzan con `turtle`. 

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> |Intenta pulsar `TAB` de nuevo. 

- Esto debería mostrar todos los paquetes de ROS que comienzan con `turtle`:

    `turtle_actionlib/         turtlesim/         turtle_tf/`

- En la línea de comandos, deberías tener de nuevo: `$ roscd turtle`

- Ahora escribe una s después de `turtle` y luego presiona TAB:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> | `$ roscd turtles<<< ahora pulse la tecla TAB >>>>`

- Como sólo hay un paquete que comienza con `turtels`, deberías ver: `$ roscd turtlesim/`

- Si deseas ver una lista de todos los paquetes instalados actualmente, también puede usar la compleción con TAB:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> | `$ rosls <<< ahora pulse la tecla TAB >>>>`
