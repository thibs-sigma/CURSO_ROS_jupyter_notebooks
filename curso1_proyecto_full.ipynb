### Desarollo del script

- Empeza la creación de un nuevo Python script, llamado `laser_distance.py`:

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/Tango icon library/Apps/accessories-text-editor.svg"> | `$ roscd test_package/src` <br> `$ gedit laser_distance.py`

- El editor de texto se debe abrir. Copia/Pega el código siguiente (no te asuste, vamos a explicar todo más adelante):

 ```python 
 #!/usr/bin/env python

 import rospy
 from math import sin, cos
 
 from sensor_msgs.msg import LaserScan

 # Initialization
 laser_data = None

 min_object = 355 # Min object detection
 max_object = 365 # Max object detection

 def callbackFunctionForLaser(data):
     global laser_data
     laser_data = data
     print('Data received: ', laser_data.ranges[360])

     location = (max_object + min_object)/2
     distance = (laser_data.ranges[max_object] + laser_data.ranges[min_object])*1000/2
     location =  location * laser_data.angle_increment + laser_data.angle_min

     pos_x = sin(location) * distance
     pos_y = cos(location) * distance
     pos_z = 2

     print ('position x:', pos_x)
     print ('position y:', pos_y)
     print ('position z:', pos_z)

 if __name__ == '__main__':
     try:
         rospy.init_node('laser_distance')

         # Callback routines
         rospy.Subscriber("/ur3/laser/scan", LaserScan, callbackFunctionForLaser) # Laser subscriber

         rate = rospy.Rate(50)

         # Debug terminal
         print "Ready to go!"

         rospy.spin()

     except rospy.ROSInterruptException:
         pass
 ```

- Guarda los modificaciones (`CTRL+S` o Archivo > Guardar)

### Explicaciones

- Vamos a descifrar el código. 

#### Inicializaciónes e imports

```python
#!/usr/bin/env python
```

- Cada Nodo ROS de Python tendrá esta declaración en la parte superior. La primera línea se asegura de que el script se ejecute como un script Python. 

```python
import rospy
from math import sin, cos
from sensor_msgs.msg import LaserScan
```

- Necesita importar `rospy` si está escribiendo un nodo ROS (es nuestro caso acá). 

- También necesitamos las funciones matemáticas `sin` y `cos` para la computación de las distancias.

- La importación `sensor_msgs.msg` es para que podamos reutilizar el tipo de mensaje `sensor_msgs/LaserScan` para su publicación. 

```python
# Initialization
laser_data = None

min_object = 355 # Min object detection
max_object = 365 # Max object detection
```

- Esos son inicializaciónes de variables de Python

#### Main función

```python
rospy.init_node('laser_distance')
```

- Ese comando es muy importante ya que le dice a `rospy` el nombre del nodo.

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/important-red.svg"> | Hasta que rospy tenga esta información, no puede empezar a comunicarse con el *ROS Master*. En este caso, el nodo tendrá el nombre `laser_distance`

```python
rospy.Subscriber("/ur3/laser/scan", LaserScan, callbackFunctionForLaser)
```

- Esto declara que el nodo se suscribe al *Topic* del Laser `/ur3/laser/scan` que es de tipo `sensor_msgs.msgs.LaserScan`. Cuando se reciben nuevos mensajes, la función `callbackFunctionForLaser` se invoca con el mensaje como primer argumento. 

```python
rate = rospy.Rate(50)
```

- Esta línea crea una un objeto `rate`. Con la ayuda de su método `sleep()`, ofrece una forma conveniente de hacer bucles a la velocidad deseada. Con su argumento de 10, deberíamos esperar pasar por el bucle 10 veces por segundo (siempre y cuando nuestro tiempo de procesamiento no supere la décima parte de un segundo). 

```python
rospy.spin()
```

- `rospy.spin()` simplemente evita que su nodo salga hasta que el nodo se haya apagado.

```python
try:
    ...
except rospy.ROSInterruptException:
    pass
```

- Además de la comprobación estándar de Python `__main__`, esto atrapa una excepción `rospy.ROSInterruptException`, que puede ser lanzada por los métodos `rospy.sleep()` y `rospy.Rate.sleep()` cuando se presiona `Ctrl-C` o cuando su Nodo se apaga de otra manera. La razón por la que se plantea esta excepción es para que no continúe ejecutando código accidentalmente después de la suspensión `sleep()`. 

#### Sensor a cartesiano

- Deseamos calcular la posición del objeto en el mundo del robot.

- Así, necesita hacer una conversión entre los datos del sensor hacía las posiciones. 

>| []() |  []() |
|:---:|:---:| 
| <img src="IMAGES/objects/notice.svg"> | Podríamos también hacer la conversión entre las posiciones hacía los comandos de las articulaciones. Pero tomaría demasiado tiempo en este curso, así que decidio presentarles los conceptos separados.

- Del conjunto de medidas de distancia podemos encontrar los objetos alrededor del brazo.

- Cada objeto está representado por múltiples mediciones. 

- Cualquier medición que devuelve un valor entre el valor mínimo y el valor máximo ha encontrado un objeto. 

<img src="IMAGES/explication_laser.png">

- Dado que los objetos son redondos (latas de Coca-Cola), podemos tomar el centro de cada objeto como punto a apuntar. 

- La distancia hasta este punto puede calcularse como el promedio entre la distancia observada más a la derecha y la distancia observada más a la izquierda.

- Así, escribimos las lineas siguientes:

```python
location = (max_object + min_object)/2
distance = (laser_data.ranges[max_object] + laser_data.ranges[min_object])*1000/2
location =  location * laser_data.angle_increment + laser_data.angle_min
```

- Luego, tomando el sinus del ángulo de la muestra y multiplicándolo por la distancia, se obtiene la coordenada x. 

- El mismo procedimiento con el cosinus nos daría la coordenada y. 

- Cualquier valor entre 0 y la altura del objeto sería aceptable para la coordenada z.

```python
pos_x = sin(location) * distance
pos_y = cos(location) * distance
pos_z = 2
```

#!/usr/bin/env python

import rospy
import math

from std_msgs.msg import Float64
from math import sin, cos, atan2, sqrt, fabs

# Define publishers for joint1 and joint2 position controller commands.
joint1publisher = rospy.Publisher('/rrbot/joint1_position_controller/command', Float64, queue_size=10)
joint2publisher = rospy.Publisher('/rrbot/joint2_position_controller/command', Float64, queue_size=10)


# RRArm initial joint positions publisher for joint controllers.
def rrarm_positions_publisher():
    global joint1publisher, joint2publisher
    positioning = True
    joint1initial_position = 1.0
    joint2initial_position = 2.4
    move_step_joint1 = 0.01
    move_step_joint2 = 0.035
    step_upward1 = 0.02
    step_upward2 = 0.04

    joint1current_position = joint1initial_position
    joint2current_position = joint2initial_position

    # Initialization of node for controlling joint1 and joint2 positions.
    rospy.init_node('joint_positions_node', anonymous=True)

    rate = rospy.Rate(80)  # Rate 80 Hz

    while not rospy.is_shutdown() and positioning:
        raw_input("Press Enter to go to the initial position...")

        
        rospy.logdebug('Going to initial position...')
        rrarm_initial_joints_position(joint1initial_position, joint2initial_position)
        
        # Wait 1 sec
        rospy.sleep(1)
        
        raw_input("Press Enter to continue and push the object...")

        rospy.logdebug('Calling the function to push the object...')
        joint1current_position, joint2current_position = rrarm_pushing_object(joint1initial_position,
                                                                              joint2initial_position, move_step_joint1,
                                                                              move_step_joint2)
        # Wait 1 sec
        rospy.sleep(1)

        raw_input("Press Enter to continue and roll back the arm...")

        rospy.logdebug('Rolling back the arm...')
        joint1current_position, joint2current_position = rrarm_roll_back(joint1current_position, joint2current_position,
                                                                         step_upward1, step_upward2)
        
        #Wait 1 sec
        rospy.sleep(1)

        raw_input("Press Enter to continue and lift up the arm...")

        rospy.logdebug('Lifting up the arm...')
        joint1current_position = rrarm_lift_up(joint1current_position, step_upward1)

        positioning = False
    
    rospy.logdebug('ACTION COMPLETED! Exiting...')


def rrarm_initial_joints_position(joint1initial_position, joint2initial_position):
    publish_joints_position(joint1initial_position, joint2initial_position)
    rospy.loginfo('joint1 position: %s', joint1initial_position)
    rospy.loginfo('joint2 position: %s', joint2initial_position)


def rrarm_pushing_object(joint1initial_position, joint2initial_position, move_step_joint1, move_step_joint2):
    for x in range(0, 50):
        joint1_position = joint1initial_position + move_step_joint1 * x
        joint2_position = joint2initial_position - move_step_joint2 * x
        publish_joints_position(joint1_position, joint2_position)

        # Debug
        rospy.loginfo('Step %s / 50', x+1)
        rospy.loginfo('joint2 position: %s', joint2_position)
        
        # Smoothen the movement
        rospy.sleep(0.05)

    return joint1_position, joint2_position


def rrarm_roll_back(joint1_position, joint2_position, step_upward1, step_upward2):
    for x in range(1, 10):
        joint2_position = joint2_position + step_upward2
        joint1_position = joint1_position - step_upward1
        publish_joints_position(joint1_position, joint2_position)
        
        # Debug
        rospy.loginfo('Step %s / 9', x)
        rospy.loginfo('joint1 position: %s', joint1_position)
        
        # Smoothen the movement
        rospy.sleep(0.05)

    return joint1_position, joint2_position


def rrarm_lift_up(joint1_position, step_upward1):
    for x in range(1, 10):
        joint1_position = joint1_position - step_upward1 * x

        publish_joint_position(joint1_position, 1)
        
        # Debug
        rospy.loginfo('Step %s / 9', x)
        rospy.loginfo('joint1 position: %s', joint1_position)

        # Smoothen the movement
        rospy.sleep(0.05)

    return joint1_position


def publish_joints_position(joint1_position, joint2_position):
    joint1publisher.publish(joint1_position)
    joint2publisher.publish(joint2_position)


def publish_joint_position(joint_position, joint_number):
    if joint_number == 1:
        joint1publisher.publish(joint_position)
    elif joint_number == 2:
        joint2publisher.publish(joint_position)


# Below code that will continuously run (to stop it press CTRL+C)
if __name__ == '__main__':
    try:
        rrarm_positions_publisher()
    except rospy.ROSInterruptException:
        pass

```python
#!/usr/bin/env python

import rospy, sys
import moveit_commander

from std_msgs.msg import Bool
from moveit_msgs.msg import RobotTrajectory, PlanningScene
from trajectory_msgs.msg import JointTrajectoryPoint

from geometry_msgs.msg import PoseStamped, Pose

import tf


def UR3_motionPlanning():
    # Initialize the move_group API
    moveit_commander.roscpp_initialize(sys.argv)
    
    # Initialize the move group for the right arm
    arm = moveit_commander.MoveGroupCommander('manipulator')
            
    # Get the name of the end-effector link
    end_effector_link = arm.get_end_effector_link()
    rospy.loginfo("End effector link: %s", end_effector_link)
                    
    # Set the reference frame for pose targets
    reference_frame = '/base_link'
    
    # Set the right arm reference frame accordingly
    arm.set_pose_reference_frame(reference_frame)
            
    # Allow replanning to increase the odds of a solution
    arm.allow_replanning(True)
    
    # Allow some leeway in position (meters) and orientation (radians)
    arm.set_goal_position_tolerance(0.05) # Changed from 0.01 to 0.05
    arm.set_goal_orientation_tolerance(0.05)

    # Set velocity scaling factor
    arm.set_max_velocity_scaling_factor(0.3)
    arm.set_max_acceleration_scaling_factor(0.3)

	#Debug
    rospy.loginfo(arm.get_current_pose(end_effector_link))
    rospy.logwarn("Waiting 5 sec")
    rospy.sleep(5)            
            
    # Set the target pose
    target_pose = PoseStamped()
    target_pose.header.frame_id = reference_frame
    target_pose.header.stamp = rospy.Time.now()   

    # Test for UR3
    target_pose.pose.position.x = 0.464
    target_pose.pose.position.y = 0.180
    target_pose.pose.position.z = 0.281

    target_pose.pose.orientation.x = -0.479
    target_pose.pose.orientation.y = 0.514
    target_pose.pose.orientation.z = 0.505
    target_pose.pose.orientation.w = 0.501

    # Set the start state to the current state
    arm.set_start_state_to_current_state()
    
    # Set the goal pose of the end effector to the stored pose
    arm.set_pose_target(target_pose, end_effector_link)
    
    # Plan the trajectory to the goal
    traj = arm.plan()
    
    # Execute the planned trajectory
    rospy.logwarn("Executing target pose (IK)")
    arm.execute(traj)

    # Pause for a second
    rospy.logwarn("Waiting 5 sec")
    rospy.sleep(5)

    # Shut down MoveIt cleanly
    moveit_commander.roscpp_shutdown()
    
    # Exit MoveIt
    moveit_commander.os._exit(0)

def UR3_setup():
    # Initialize the move_group API
    moveit_commander.roscpp_initialize(sys.argv)

    # Use the planning scene object to add or remove objects //Interface
    scene = moveit_commander.PlanningSceneInterface()

    rospy.sleep(2)

    REFERENCE_FRAME = '/world'
    p = PlanningScene()
    p.is_diff = True    

    # Create a scene publisher to push changes to the scene //PlanningScene
    scene_pub = rospy.Publisher('/move_group/monitored_planning_scene', PlanningScene, queue_size=1)

    pose_Ground = PoseStamped()
    pose_Ground.header.frame_id = REFERENCE_FRAME
    pose_Ground.pose.position.x = 0.0
    pose_Ground.pose.position.y = 0.0
    pose_Ground.pose.position.z = 0.0
    pose_Ground.pose.orientation.w = 1.0

    # Remove leftover objects from a previous run
    scene.remove_world_object('ground')

	# Add ground to the planning scene
    scene.add_box("ground", pose_Ground, (1.0, 1.0, 0.001))
    
    # Publish the Planning Scene    
    scene_pub.publish(PlanningScene)

    # Initialize the move group for the right arm
    arm = moveit_commander.MoveGroupCommander('manipulator')
            
    # Get the name of the end-effector link
    end_effector_link = arm.get_end_effector_link()
    rospy.loginfo("End effector link: %s", end_effector_link)
                    
    # Set the reference frame for pose targets
    reference_frame = '/base_link'
    
    # Set the right arm reference frame accordingly
    arm.set_pose_reference_frame(reference_frame)
            
    # Allow replanning to increase the odds of a solution
    arm.allow_replanning(True)
    
    # Allow some leeway in position (meters) and orientation (radians)
    arm.set_goal_position_tolerance(0.05) # Changed from 0.01 to 0.05
    arm.set_goal_orientation_tolerance(0.05)

    # Set velocity scaling factor
    arm.set_max_velocity_scaling_factor(0.3)
    arm.set_max_acceleration_scaling_factor(0.3)

    default_joint_states = arm.get_current_joint_values()

    # Define joint_states
    default_joint_states[0] = 0.1388552188873291
    default_joint_states[1] = -0.569965124130249
    default_joint_states[2] = 0.17023944854736328
    default_joint_states[3] = -1.1359660625457764
    default_joint_states[4] = -1.5523438453674316
    default_joint_states[5] = -1.4006805419921875

    # Initial position
    arm.set_joint_value_target(default_joint_states)

    # Set the internal state to the current state
    arm.set_start_state_to_current_state()
    plan = arm.plan()

	# Execute motion planning
    arm.execute(plan)


if __name__ == "__main__":
    try:
		rospy.init_node('moveit_demo')
		
		# ROS stuff
		rate = rospy.Rate(50)

		rospy.loginfo("Moving to setup pose...")

		UR3_setup()

		raw_input("Press Enter to continue and execute the motion planning...")

		UR3_motionPlanning()
		
		rospy.spin()
    
    except rospy.ROSInterruptException:
        pass
```
